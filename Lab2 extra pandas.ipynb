```python
# Code for doing the above thing for all the authors
def get_author(url):
    name = url.split('/')[-1].split('.')[1:][0]
    return name
```

```python
# apply get_author to author_url using .map
# add new column author 
df['author'] = df.author_url.map(get_author)
df.author[0:5]
```

```python
# Parse out genres form genre_urls
test_genre_string = df.genre_urls[0]
genres = test_genre_string.strip().split('|')
for e in genres:
    print(e.split('/')[-1])
    "|".join(genres)
```

```python
# accepts a genre url and returns the genre name
def split_and_join_genres(url):
    genres = url.strip().split('|')
    genres = [e.split('/')[-1] for e in genres]
    return "|".join(genres)
```

```python
#use map to create new 'genre' columns
df['genres'] = df.genre_urls.map(split_and_join_genres)
df.head()
```
